## Object Oriented API (More flexible than pyplot)

```python
# 1st Method
fig = plt.figure() # Creates a figure
ax = fig.add_subplot() # Add some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
# ax = fig.add_subplot() # Add some axes
ax.plot(x, y); #add some data
# plt.show()
```

```python
# 3rd Method ** Recommended**

fig, ax = plt.subplots()
ax.plot(x, y);
type(fig), type(ax)
```

## Matplotlib Workflow

```python
# 0. Import matplot lib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 12, 13, 14]

# 3 Setup plot
fig, ax = plt.subplots(figsize=(5, 5))

# 3. Plot data
ax.plot(x, y)

# 4 Customize plot
ax.set(title=" Simple Plot",
      xlabel="X- Axis",
      ylabel="Y-Axis")

# 5 Save & show (you save the whole figure)
fig.savefig("plot-1.png")
```

```python
# Make a plot from a Dictionary
nut_butter_prices = {
    "Almond Butter": 10,
    "Peanut Butter": 15,
    "Cashew: Butter": 4
}

fig, ax = plt.subplots();
ax.set(title="Prices of nut butters", xlabel="Butters", ylabel="Prices")
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Prices of nut butters in horizontal manner graph");
```

## Two Options for subplots

### Option 1 

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10, 5))
# Plot each axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

### Option 2

```python
## Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(20), np.random.random(20))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000)); 
```

# Which one should we use? (pyplot vs matplotlib OO method)
* pyplot - Okay when plotting something quickly
* OO Method - When plotting something advanced

```python
over_50 = heart_disease[heart_disease["age"] > 50]
# len(over_50)
over_50
```

```python
over_50.head()
```

```python
# pyplot method
over_50.plot(x="age", 
             y="chol", 
             kind="scatter", 
             c="target");
```

```python
## OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(x="age", 
             y="chol", 
             kind="scatter", 
             c="target",
            ax=ax);
# ax.set_xlim([45, 100])
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))
# plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Adding labels and title
ax.set(title="Heart Diseases & Cholesterol Levels", 
       xlabel="Age", 
       ylabel="Cholesterol");

# Adding legends
ax.legend(*scatter.legend_elements(), title="Target")
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
## Subplots of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10, 10),
                             sharex=True)
# Add data to ax0
ax0.scatter(x=over_50["age"], 
            y=over_50["chol"], 
            c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterols Levels",
       ylabel="Cholestrol")

# Add legends to ax0
ax0.legend(*scatter.legend_elements(),
           title="Target")

# Add mean ine to ax0
ax0.axhline(over_50["chol"].mean(), 
           linestyle="--")

# Add data to ax1
ax1.scatter(x=over_50["age"],
          y=over_50["thalach"],
          c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age", # The whole figure shares this label across axeses
       ylabel="Max Heart Rate")

# Add legend to ax1
ax1.legend(*scatter.legend_elements(),
          title="Target")

ax1.axhline(over_50["thalach"].mean(),
           linestyle="--")

# Add title to figure
fig.suptitle("Heart Disease Analysis", fontsize="16", fontweight="bold");

```
