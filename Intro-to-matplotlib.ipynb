## Object Oriented API (More flexible than pyplot)

```python
# 1st Method
fig = plt.figure() # Creates a figure
ax = fig.add_subplot() # Add some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
# ax = fig.add_subplot() # Add some axes
ax.plot(x, y); #add some data
# plt.show()
```

```python
# 3rd Method ** Recommended**

fig, ax = plt.subplots()
ax.plot(x, y);
type(fig), type(ax)
```

## Matplotlib Workflow

```python
# 0. Import matplot lib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 12, 13, 14]

# 3 Setup plot
fig, ax = plt.subplots(figsize=(5, 5))

# 3. Plot data
ax.plot(x, y)

# 4 Customize plot
ax.set(title=" Simple Plot",
      xlabel="X- Axis",
      ylabel="Y-Axis")

# 5 Save & show (you save the whole figure)
fig.savefig("plot-1.png")
```

## Making figure with Numpy Arrays

We are creating:
    Line Plot
    Scatter Plot
    Bar Plot
    Histogram
    Subplot

```python
# Generate data
x = np.linspace(1, 10, 100)
x[:10]
```

```python
# Create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to create scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from a Dictionary
nut_butter_prices = {
    "Almond Butter": 10,
    "Peanut Butter": 15,
    "Cashew: Butter": 4
}

fig, ax = plt.subplots();
ax.set(title="Prices of nut butters", xlabel="Butters", ylabel="Price")
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
```
